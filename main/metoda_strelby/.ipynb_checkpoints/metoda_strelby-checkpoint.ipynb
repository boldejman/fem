```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
```

```python
a, b, A, B = 0, 5, 0, 1
```

```python
def rhs(x):
    return np.sin(x)-x*np.cos(2*x)
```

```python
def euler_method(v, h: int = 0.0001, only_ans=True):
    u_i = A
    v_i = v
    x_i = 0
    u = np.zeros(int(5/h))
    for i in range(int(5/h)):
        x_i += h
        u_i += h*v_i
        v_i += h*rhs(x_i)
        u[i] = u_i
    if only_ans:
        return u[-1] - B
    return u

v_t = newton_method(euler_method, -10, 10)
u_h = euler_method(v_t, only_ans = False)
```

```python
x = np.linspace(0, 5, int(5/0.0001))
y = u_h
u_ex = (4 * x + x * np.sin(10) + 4 * x * np.sin(5) - 20 * np.sin(x) - 5 * np.sin(2*x) + 5 * x * np.cos(2*x)- 5 * x * np.cos(10)) / 20
# plt.plot(x, y)
# plt.plot(x, u_ex)
plt.plot(x, u_ex - y)
print(y[-1])
```

```python
def bisection_method(f, x_a, x_b, tol=0.000001, max_iter=100) -> float:
    a, b = x_a, x_b
    f_a, f_b = f(a), f(b)
    
    if f_a * f_b > 0:
        raise Exception("bad interval")
        
    if f_a == 0. or abs(f_a) < tol:
        return a
    
    if f_b == 0. or abs(f_b) < tol:
        return b
        
    n = 1
    while n < max_iter:
        c = (a + b) / 2
        f_c = f(c)
        if f_c == 0.0 or abs((b-a)/2) < tol:
            return c
        else:
            n += 1
            if f_c * f(a) > 0:
                a = c
            else:
                b = c
    raise Exception("max iterations exceed with f={0}".format(f_c))
    
```

```python
def newton_method(f, x_0, x_b, tol=0.000001, max_iter=1000) -> float:
    def df(f, x, h=tol):
        return (f(x+h) - f(x))/h
    
    n = 1
    while n < max_iter:
        y = f(x_0)
        dy = df(f, x_0)
        x_1 = x_0 - (y / dy)
        if abs(x_1 - x_0) < tol:
            return x_1
        
        x_0 = x_1
        n += 1
        
    raise Exception("max iterations exceed with f={0}".format(f(x_0)))
    
print(newton_method(rhs, 3, 0), bisection_method(rhs, 2, 3))

        
```
